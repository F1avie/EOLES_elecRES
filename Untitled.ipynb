## On teste la contrainte

```python
#On créé un Data Frame sur lequel on va tester les ocntraintes
test_contrainte = simu_opti_ademe
test_contrainte
```

```python
#ajout de la production nette pour chaque heure
ProdTotale = test_contrainte.onshore + test_contrainte.offshore + test_contrainte.pv +\
test_contrainte.river + test_contrainte.lake + test_contrainte.biogas #+ A.phs + A.battery + A.methanation

delta = ProdTotale - test_contrainte['Electricity demand']
test_contrainte['production_nette'] = delta
```

```python
#ces colonnes ne sont pas utilisées
test_contrainte.drop(['onshore', 'offshore', 'pv', 'river','lake', 'biogas', 'phs', 'battery', 'methanation'],inplace = True, axis =1)
```

```python
#delta_stock heure par heure
next_hour = test_contrainte[1:][['Stored phs','Stored battery','Stored methanation']]
next_hour.columns = ['Stored_H+1_phs','Stored_H+1_battery','Stored_H+1_methanation']
next_hour = next_hour.reset_index(drop=True)
```

```python
test_contrainte = test_contrainte[:-1]
test_contrainte = [test_contrainte,next_hour]
test_contrainte =pd.concat(test_contrainte,axis =1)
```

```python
#création des Delta Stock pour chaque heure
test_contrainte ['stock_hour_phs'] = test_contrainte['Stored_H+1_phs'] - test_contrainte['Stored phs']
test_contrainte ['stock_hour_battery'] = test_contrainte['Stored_H+1_battery'] - test_contrainte['Stored battery']
test_contrainte ['stock_hour_methanation'] = test_contrainte['Stored_H+1_methanation'] - test_contrainte['Stored methanation']
```

```python
#On supprime les lignes correspondants au passage d'une année à une autre
test_contrainte["hours_year"]= test_contrainte.index %(365*24) +1
test_contrainte.drop(test_contrainte[test_contrainte.hours_year == (365*24)].index, axis = 0, inplace = True)
```

```python
def C_residualprod(RP,eta_in, eta_out, delta_stock):
    ################################################
    # RP is the residual production : production_nette
    # eta_in Charging efficiency of phs (0), battery (1), methanation (2)
    # eta_out Disharging efficiency of phs (0), battery (1), methanation (2)
    # delta_stock = variation des stocks dans la journée
    # volume = volume max of each tec (useful to unpercent)
    
    # return index of problem
    ################################################   
    
    deltaphs = delta_stock[0]#*volume[0]/100
    deltabat = delta_stock[1]#*volume[1]/100
    deltamet = delta_stock[2]#*volume[2]/100
    #if delta_stock > 0
    deltaphs[deltaphs>0] = deltaphs[deltaphs>0]/eta_in[0]
    deltabat[deltabat>0] = deltabat[deltabat>0]/eta_in[1]
    deltamet[deltamet>0] = deltamet[deltamet>0]/eta_in[2]
    # if delta_stock < 0

    
    deltaphs[deltaphs<0] = deltaphs[deltaphs<0]*eta_out[0]
    deltabat[deltabat<0] = deltabat[deltabat<0]*eta_out[1]
    deltamet[deltamet<0] = deltamet[deltamet<0]*eta_out[2]

    SUM = deltaphs + deltabat + deltamet
    
    # in all cases : each day, RP have to be lower than sum
    
    print('here there is a problem : ', RP[RP<SUM ], SUM[RP[RP<SUM].index])
    
    return  RP[RP<SUM], SUM[RP[RP<SUM].index],RP,SUM
```

```python
RP,SUM,RPViz,SUMViz = C_residualprod(test_contrainte.production_nette,eta_in, eta_out, [test_contrainte.stock_hour_phs, test_contrainte.stock_hour_battery, test_contrainte.stock_hour_methanation])
```

```python
start =0
finish = 11
fig, ax =plt.subplots(1, figsize = (15,5))
ax.plot(np.abs(RPViz[start:finish]-SUMViz[start:finish]),label = "Différence absolue entre la prod résiduelle et le delta de stockage")
# ax.plot(RP[start:finish], label ="Delta Stock")
# ax.plot(SUM[start:finish], label ="Delta Stock")
X= np.array(RP.index)
X = np.unique(np.clip(start,finish,X))
for i in (X):
    ax.axvline(i, color='r',alpha=0.3)
ax.set_yscale('log')
ax.legend(loc ="upper right")
```
